```python
import nltk

nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

def analyze_sentiment(text):
    sentiment_analyzer = SentimentIntensityAnalyzer()
    sentiment = sentiment_analyzer.polarity_scores(text)
    if sentiment['compound'] > 0.05:
        return "Positive"
    elif sentiment['compound'] < -0.05:
        return "Negative"
    else:
        return "Neutral"
```

```python
test_cases = [
    "I love pizza",
    "I hate pizza",
    "Sometimes I eat pizza, sometimes I don't",
    "I love pizza, but I can't have it every day",
    "I hate pizza but I like pepperonni",
    "I love Pizza Hut but I hate Domino's",
    "I like the Hawaiian Pizza from Pizza Hut but I hate the one from Little Ceszar",
    "I had a fantastic time at the beach yesterday!",
    "I'm so disappointed with my exam results.",
    "The weather today is so beautiful and sunny.",
    "I just found out that I lost my job.",
    "I'm so grateful for my friends and family.",
    "I'm not sure if I'm ready for this new challenge.",
    "The concert was amazing and the singer was incredible.",
    "I had a terrible experience at the restaurant last night.",
    "I'm so excited for my trip to Europe next week.",
    "I feel so overwhelmed and stressed out right now.",
    "The new movie was a complete waste of time.",
    "I love spending time with my pets.",
    "I'm so proud of my team for winning the championship.",
    "I'm feeling really down and depressed today.",
    "The new book I just read was so captivating and inspiring.",
    "I'm so frustrated with this traffic jam and it's taking forever to get home.",
    "I had the best vacation ever and I can't wait to go back.",
    "I'm so angry at my boss for treating me unfairly.",
    "The new restaurant in town has amazing food and service.",
    "I'm feeling really hopeful and optimistic about the future.",
]
for t in test_cases:
    print(t)
    sentiment = analyze_sentiment(t)
    print("The sentiment of the text is:", sentiment)
    print("")
```
